```python
## Creating Data
df = [list(range(1,51)), list(range(101,151)), list(range(501,551))]
print(df)
```

```python
## Printing Data
print([row[:5] for row in df])
```

```python
## Splitting Data into Train and Test Sets with Shuffling in Lists in Python

import random
random.seed(0)
# Empty lists for features and labels and indexes
indexes = []
X=[ [], [] ]
y = []
# Shuffle indexes
indexes = list(range(len(df[0])))
random.shuffle(indexes)
# Create X and y based on shuffled indexes
for i in range(len(indexes)):
    X[0].append(df[0][indexes[i]]) # First column is first feature
    X[1].append(df[1][indexes[i]]) # Second column is second feature
    y.append(df[-1][indexes[i]]) # Last column is labels
# Define a split point
split_point = int(0.8*len(indexes))
# Split the data into training and testing sets
X_train =  [X[0][:split_point], X[1][:split_point]]
X_test =  [X[0][split_point:], X[1][split_point:]]
y_train = y[:split_point]
y_test = y[split_point:]
```

```python
# Check lengths of the splits
print(len(X_train[0]),len(X_test[0]),len(y_train),len(y_test))
```
